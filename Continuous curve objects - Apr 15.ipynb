# Defining continuous curves in Python

A continuous curve is a topological object defined by the *graph* of a continuous function. Let $I \subset \mathbb{R}$ be a real interval and $X$ be a topological space. The continuous curve is given by
$$
    \gamma : I \rightarrow X.
$$

I'm going to start with simple conic sections. A generalized conic is described by the equation
$$
    r = \frac{r_0}{1 + \lambda \sin(k\theta)},
$$
where $r_0 \geq 0$, $\lambda \geq 0$, and $k \in \mathbb{R}$. The conic is respectively called a generalized hyperbola, parabola, or ellipse if $\lambda > 1$, $\lambda = 1$, or $\lambda < 1$

```python
import numpy as np
import matplotlib.pyplot as plt

class Conic:
    def __init__(self, r0, lmbda, k, theta):
        self.r0 = r0
        self.lmbda = lmbda
        self.k = k
        self.theta = theta
        
        self.radius = r0/(1 + lmbda*np.sin(k*theta))
        
        if self.lmbda > 1:
            self.name = 'hyperbola'
        elif self.lmbda < 1:
            self.name = 'ellipse'
        else:
            self.name = 'parabola'
        
    def display(self, kind = 'p'):
        print('Displaying the plot of a generalized {}'.format(self.name))
        
        if kind == 'c':
            plt.plot(self.theta, self.radius);
        elif kind == 'p':
            plt.polar(self.theta, self.radius)
        else:
            print('Unknown argument "kind". Displaying in polar coordinates.')
            plt.polar(self.theta, self.radius)
```

```python
c1 = Conic(1, 2.568, 1., np.linspace(0, 2*np.pi, 101))
```

```python
c1.display()
```
