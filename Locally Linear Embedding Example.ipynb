```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
```

```python
X, _ = load_digits(return_X_y = True)
X.shape
```

```python
plt.imshow(X[0].reshape((8,8)), cmap = 'gray');
```

Ok, so the data is made up of 8x8 pixel images of digits, like a baby MNIST.

```python
# Initialize the model for the embedding with a manifold dimension of k
k = 4

embedding = LocallyLinearEmbedding(n_components = k)
```

```python
help(embedding)
```

```python
W = embedding.fit(X[:100])
W.embedding_.shape
```

```python
W.embedding_
```

```python
# Fit the parameters for the first 100 data points
Y = embedding.fit_transform(X[:100])
```
