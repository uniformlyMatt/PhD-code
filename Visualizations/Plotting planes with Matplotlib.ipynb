```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

point  = np.array([1, 2, 3])
normal = np.array([1, 1, 2])

n1 = np.random.normal(0, 1, 3)
n2 = np.random.normal(0, 1, 3)
n3 = np.random.normal(0, 1, 3)

p1 = np.array([1, 0, 1])
p2 = np.array([1, 1, 0])
p3 = np.array([0, 1, 1])

d11 = -p1.dot(n1)
d12 = -p1.dot(n2)
d13 = -p1.dot(n3)

d21 = -p2.dot(n1)
d22 = -p2.dot(n2)
d23 = -p2.dot(n3)

d31 = -p3.dot(n1)
d32 = -p3.dot(n2)
d33 = -p3.dot(n3)

# a plane is a*x+b*y+c*z+d=0
# [a,b,c] is the normal. Thus, we have to calculate
# d and we're set
d = -point.dot(normal)

# create x,y
xx, yy = np.meshgrid(range(10), range(10))

# calculate corresponding z
z = (-normal[0] * xx - normal[1] * yy - d) * 1. /normal[2]

z11 = (-n1[0]*xx - n1[1]*yy - d11)*1./n1[2]
z12 = (-n1[0]*xx - n1[1]*yy - d12)*1./n1[2]
z13 = (-n1[0]*xx - n1[1]*yy - d13)*1./n1[2]
z21 = (-n2[0]*xx - n2[1]*yy - d21)*1./n2[2]
z22 = (-n2[0]*xx - n2[1]*yy - d22)*1./n2[2]
z23 = (-n2[0]*xx - n2[1]*yy - d23)*1./n2[2]
z31 = (-n3[0]*xx - n3[1]*yy - d31)*1./n3[2]
z32 = (-n3[0]*xx - n3[1]*yy - d32)*1./n3[2]
z33 = (-n3[0]*xx - n3[1]*yy - d33)*1./n3[2]

# plot the surface
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, z11)
plt3d.plot_surface(xx, yy, z12)
plt3d.plot_surface(xx, yy, z13)

plt3d.plot_surface(xx, yy, z21)
plt3d.plot_surface(xx, yy, z22)
plt3d.plot_surface(xx, yy, z23)

plt3d.plot_surface(xx, yy, z31)
plt3d.plot_surface(xx, yy, z32)
plt3d.plot_surface(xx, yy, z33)
plt.show()
```

```python
def plane_plotter(normal, point):
    """ Function to plot a plane in 3d 
    
        Args:
        -----
        normal := array containing the normal of the plane
        point  := array of a point on the plane
        
        Returns:
        --------
        array of function values over the 20x20 grid
        
        TODO: Make the grid adjustable
    """
    
    xx, yy = np.meshgrid(np.linspace(-10,10,11), np.linspace(-10,10,11))
    
    d = -point.dot(normal)
    
    return (-normal[0] * xx - normal[1] * yy - d) * 1. /normal[2]
```

```python
import itertools

zs = [plane_plotter(normal,point) for normal,point in itertools.product([n1,n2,n3],[p1,p2,p3])]
```

```python
%matplotlib notebook

plt3d = plt.figure().gca(projection='3d')

xx, yy = np.meshgrid(np.linspace(-10,10,11), np.linspace(-10,10,11))

for z in zs:
    plt3d.plot_surface(xx, yy, z)
```
